# 6 Number of daily message posting

```python
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib notebook
file_path = 'C:/Users/User/Desktop/data/channels/karaulny.csv' #Path to your csv file
df = pd.read_csv(file_path)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
message_counts = df['date'].value_counts() - 1
message_counts = message_counts.sort_index()
plt.figure(figsize=(20, 10))
plt.plot(message_counts.index, message_counts.values, color='green', linestyle='-', linewidth=1, markersize=10)
plt.title('Frequency of postingм')
plt.xlabel('Date')
plt.ylabel('Number of messages')
plt.show()


```

# 8 Diagram of daily views 

```python
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib notebook
file_path = 'C:/Users/User/Desktop/data/channels/karaulny.csv' #Path to your csv file
df = pd.read_csv(file_path)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
daily_views = df.groupby(df['date'].dt.date)['views'].sum() 
plt.figure(figsize=(14, 6))
daily_views.plot(kind='line', color='purple')
plt.title('Daily messages views')
plt.xlabel('Date')
plt.ylabel('Views')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()

```

# 11 Top list of words, that appears in this channel more frequently

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
nltk.download('stopwords') # Аби не виводити сполучники та прийменники

file_path = 'C:/Users/User/Desktop/data/channels/karaulny.csv'   #Path to your csv file
df = pd.read_csv(file_path)

def get_top_words(texts, top_n=30):
    stop_words = set(stopwords.words('russian'))  
    words = [word.lower() for text in texts.fillna('') for word in word_tokenize(str(text)) if word.isalpha() and word.lower() not in stop_words]
    freq_dist = FreqDist(words)
    return freq_dist.most_common(top_n)

top_words = get_top_words(df['message'])

print("Top-30 words in this channel:")
for word, count in top_words:
    print(f"{word}: {count}" " times")
```

# 12 division between mount of russian and english words

```python
import pandas as pd
from langdetect import detect
from nltk.tokenize import word_tokenize
import nltk
import matplotlib.pyplot as plt
nltk.download('punkt')
#%matplotlib notebook
file_path = 'C:/Users/User/Desktop/data/channels/karaulny.csv'  #Path to your csv file
df = pd.read_csv(file_path)

def detect_language(text):
    try:
        return detect(str(text))
    except:
        return None

def count_words_by_language(text, target_language):
    try:
        if detect(str(text)) == target_language:
            words = word_tokenize(str(text))
            return len(words)
    except:
        return 0

df['russian_words'] = df['message'].apply(lambda x: count_words_by_language(x, 'ru'))
df['english_words'] = df['message'].apply(lambda x: count_words_by_language(x, 'en'))

russian_words = df['russian_words'].sum()
english_words = df['english_words'].sum()
plt.figure(figsize=(6, 6))
diagram = pd.Series({"english":english_words, "russian":russian_words})
plt.pie(diagram, labels = diagram.index, autopct='%1.1f%%', startangle=90, colors=['yellow', 'skyblue'])
plt.show()
```
